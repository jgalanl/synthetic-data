```python
import pandas as pd

from autogluon.tabular import TabularPredictor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../../data/tlp/synthetic-smote.csv')
```

```python
df.head()
```

```python
predictor= TabularPredictor(
    label='ED_2Clases',
    problem_type='binary',
    eval_metric='roc_auc',
    sample_weight='balance_weight',
    path='AutogluonModels/synthetic/smote/v1'
    )
```

```python
predictor.fit(
    train_data = df,
    presets = ['high_quality'],
    time_limit = 3600,
    auto_stack = True,
    # excluded_model_types=['KNN','RF','XT', 'LR'],
    verbosity = 2
)
```

```python
predictor.fit_summary()
```

```python
X_test = pd.read_csv(
    '../../data/tlp/X_test.csv'
)
y_test = pd.read_csv(
    '../../data/tlp/y_test.csv'
)
df_test = pd.concat([X_test, y_test], axis=1)
```

```python
auxiliary_metrics = ['accuracy', 'balanced_accuracy', 'f1', 'f1_macro', 'f1_micro', 'roc_auc', 'average_precision', 'precision', 'recall', 'log_loss', 'pac_score']
```

```python
predictor.evaluate(
    df_test,
    silent = False,
    auxiliary_metrics = auxiliary_metrics
    )
```

```python
predictors = predictor.leaderboard(
    df_test,
    extra_metrics = auxiliary_metrics,
    extra_info=True,
    silent=True
)
```

```python
predictors[['model', 'roc_auc', 'accuracy', 'balanced_accuracy', 'f1', 'precision', 'recall', 'num_features']]
```

```python
predictors.to_excel(
    '../../data/tlp/predictors-smote.xlsx',
    index=False
)
```
