```python
import pandas as pd

from autogluon.tabular import TabularPredictor
from autogluon.core.metrics import make_scorer
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_curve, auc
```

```python
def pr_auc_Metric(y_true, y_pred):
    precision, recall, _ = precision_recall_curve(y_true, y_pred)
    score = auc(recall, precision)
    return score
```

```python
ag_roc_auc_scorer = make_scorer(
    name='roc_auc',
    score_func=pr_auc_Metric,
    optimum=1,
    greater_is_better=True,
    needs_proba=True
)
```

```python
TIME_LIMIT = 2 * 3600
```

```python
extra_metrics = ['roc_auc', 'accuracy', 'balanced_accuracy', 'f1', 'f1_macro', 'f1_micro', 'average_precision', 'precision', 'recall', 'log_loss', 'pac_score']
```

```python
generator = 'tvaes'
```

```python
for i in range(0, 10):
    print(f'Fitting model {i}')
    train_data = pd.read_csv(
        f'../../data/synthetic/{generator}/set_{i}.csv'
        )

    predictor= TabularPredictor(
        label='ED_2Clases',
        problem_type='binary',
        eval_metric=ag_roc_auc_scorer,
        sample_weight='balance_weight',
        path=f'AutogluonModels/{generator}/v{i}'
    )

    predictor.fit(
        train_data = train_data,
        time_limit = TIME_LIMIT,
        presets = ['high_quality'],
        auto_stack = False,
        excluded_model_types=['KNN', 'LR'],
        num_stack_levels=0,
        num_bag_folds=0,
        verbosity = 2
    )

    df_test = pd.read_csv(
        f'../../data/test/set_{i}.csv'
    )

    predictors = predictor.leaderboard(
        df_test,
        extra_metrics = extra_metrics,
        extra_info=True,
        silent=True
    )
    predictors.to_excel(
        f'../../5_results/5_2_synthetic/{generator}/predictors_set_{i}.xlsx',
        index=False
    )

    feature_importance = predictor.feature_importance(train_data)
    feature_importance.to_excel(
        f'../../5_results/5_2_synthetic/{generator}/feature_importance_set_{i}.xlsx',
        index=True
    )
```
