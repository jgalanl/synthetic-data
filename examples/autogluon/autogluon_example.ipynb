```python
import pandas as pd

from autogluon.tabular import TabularDataset, TabularPredictor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../../data/examples/healthcare-dataset-stroke-data.csv')
```

```python
df.head()
```

```python
df_train,df_test=train_test_split(df,test_size=0.4, random_state=42)
df_train.shape,df_test.shape
```

```python
test_data=df_test.drop(['stroke'],axis=1)
test_data.head()
```

```python
predictor= TabularPredictor(
    label='stroke',
    problem_type='binary',
    eval_metric='accuracy',
    path='AutogluonModels/example',
    learner_kwargs = {'positive_class':1}
    )
```

```python
predictor.fit(
    train_data = df_train,
    verbosity = 2,
    presets=['good_quality', 'optimize_for_deployment'],
    num_gpus=1
)
```

```python
predictor.fit_summary()
```

```python
predictor.evaluate(df_test)
```

```python
predictor.leaderboard(df_train, silent=True)
```

```python
predictor.feature_importance(data=df_train)
```

```python
y_pred = predictor.predict(test_data)
y_pred=pd.DataFrame(y_pred, columns=['stroke'])
```
