```python
pt_columns = [column for column in df.columns if 'PT' in column ]
pz_columns = [column for column in df.columns if 'PZ' in column ]
pdc_columns = [column for column in df.columns if 'PDC' in column]
drop_columns.extend(pt_columns)
drop_columns.extend(pz_columns)
drop_columns.extend(pdc_columns)
df.drop(columns=drop_columns, inplace=True)
```

```python
# Replace nan
df.fillna(
    {'ISC.resultadosCovid.sintomaOtroString': ''},
    inplace=True
    )

df.fillna(0.0, inplace=True)
```

```python
df.replace([np.inf, -np.inf], 0, inplace=True)
```

```python
model = GReaT(
    llm='openai-community/gpt2', 
    batch_size=16, 
    epochs=25, 
    save_steps=400000
    )
```

```python
model.fit(df)
```

```python
num_d, num_h = df['ED_2Clases'].value_counts()
num_synthetic_data = num_d - num_h
NUM_ROWS = 500
```

```python
synthetic_data_list = list()
while len(synthetic_data_list) < num_synthetic_data:
    samples = model.sample(n_samples=NUM_ROWS, gen_batch = 50, continuous_empty_limit = 10)
    for sample in samples[samples['ED_2Clases'] == 'H'].values:
        synthetic_data_list.append(sample.tolist())
        if len(synthetic_data_list) == num_synthetic_data:
            break
```

```python
synthetic_data_df = pd.DataFrame(synthetic_data_list, columns=df.columns)
```

```python
llm_synthetic_data = pd.concat([df, synthetic_data_df], ignore_index=True)
```
